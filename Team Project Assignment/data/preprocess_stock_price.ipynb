```python
# Path to the folder containing the CSV files
folder_path = "/content/drive/My Drive/Colab_Notebooks/TEAM3_TEMG4940C/data/cleaned_stock_price/"

def get_multiple_files(folder_path):
  # Get a list of CSV file paths in the folder
  file_paths = glob.glob(folder_path + '*.csv')
  print(file_paths)

  stock_price = pd.DataFrame()
  for file_path in file_paths:
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)
    df.drop("Currency", axis = 1, inplace = True)
    if (df.isnull().any().any()):
      print(file_path)
    stock_price = pd.concat([stock_price, df])

  return stock_price

stock_price = get_multiple_files(folder_path)
```

```python
stock_price = stock_price.dropna()
stock_price["PermID"] = stock_price["PermID"].astype(int)
```
