```python
credit_spread = pd.read_csv("Credit_spread.csv")
```

```python
credit_spread['Date'] = pd.to_datetime(credit_spread['Date'], format = "%Y-%m-%d")
credit_spread['Month'] = credit_spread['Date'].dt.month
credit_spread['Year'] = credit_spread['Date'].dt.year

credit_spread.drop(columns = ['Date'], inplace = True)
```

```python
credit_spread.info()
credit_spread.head(20)
```

```python
grouped = credit_spread.groupby(['RIC', 'Year', 'Month'])

preprocessed_credit_spread = []

for name, group in grouped:
    last_row = group.iloc[:1, :]
    preprocessed_credit_spread.append(last_row)

preprocessed_credit_spread = pd.concat(preprocessed_credit_spread, axis=0).reset_index(drop = True)
```

```python
preprocessed_credit_spread.info()
preprocessed_credit_spread.head(5)
```

```python
preprocessed_credit_spread = preprocessed_credit_spread.assign(Credit_spread = np.mean(preprocessed_credit_spread[['CreditSpreadBid','CreditSpreadAsk']], axis=1))

preprocessed_credit_spread = pd.concat([preprocessed_credit_spread['RIC'], preprocessed_credit_spread.loc[:, 'Month':'Credit_spread']], axis = 1)
```

```python
preprocessed_credit_spread.info()
preprocessed_credit_spread.head(5)
```

```python
credit_spread_csv = preprocessed_credit_spread.to_csv(index=False)

with open('/content/drive/My Drive/Colab_Notebooks/TEAM3_TEMG4940C/data/7_credit_spread.csv', 'w') as f:
  f.write(credit_spread_csv)
```
