```python
import glob

# Path to the folder containing the CSV files
public_folder_path = "/content/drive/My Drive/Colab_Notebooks/TEAM3_TEMG4940C/data/company_financial_ratios/public/"
private_folder_path = "/content/drive/My Drive/Colab_Notebooks/TEAM3_TEMG4940C/data/company_financial_ratios/private/"
unknown_folder_path = "/content/drive/My Drive/Colab_Notebooks/TEAM3_TEMG4940C/data/company_financial_ratios/unknown/"

def get_multiple_files(folder_path):
  # Get a list of CSV file paths in the folder
  file_paths = glob.glob(folder_path + '*.csv')
  print(file_paths)

  financials = pd.Series([])
  for file_path in file_paths:
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)

    file_name = file_path.split('/')[-1].split('.')[0]
    # Transform DataFrame into a Series
    PermID = pd.Series(np.repeat(file_name, df.shape[0]), name='PermID')

    df = pd.concat([PermID, df], axis = 1)
    financials = pd.concat([financials, df])

  return financials

public = get_multiple_files(public_folder_path)
private = get_multiple_files(private_folder_path)
unknown = get_multiple_files(unknown_folder_path)
```

```python
public = public.drop(public.columns[0], axis=1)
public.info()
public.head(5)
```

```python
private = private.drop(private.columns[0], axis=1)
private.info()
private.head(5)
```

```python
unknown = unknown.drop(unknown.columns[0], axis=1)
unknown.info()
unknown.head(5)
```

```python
financials = pd.concat([public, private, unknown], axis = 0)
bonds = pd.read_csv('Dataset_1.csv')
```

```python
print(financials['PermID'].unique().shape[0])
print(bonds['PermID'].unique().shape[0])

financials.info()
financials.head(5)
```

```python
common = np.intersect1d(financials['PermID'].astype(int).unique(), bonds['PermID'].unique())
financials = financials[financials['PermID'].astype(int).isin(common)]

financials['PermID'] = financials['PermID'].astype(int)
```

```python
print(financials['PermID'].unique().shape[0])
print(financials['PermID'].unique())
financials = financials.drop(financials.columns[-1], axis = 1)
print(financials.shape)
```

```python
previous_financials = pd.read_csv('data/5_financials.csv')
previous_financials = previous_financials.dropna(subset = "Tier 1 Capital Ratio")

previous_financials.info()
previous_financials.head(20)
```

```python
common_financials = np.intersect1d(financials['PermID'].astype(int).unique(), previous_financials['PermID'].astype(int).unique())
print(common_financials)

financials = financials[~financials['PermID'].isin(common_financials)]
print(financials['PermID'].unique().shape[0])
```

```python
print(financials.shape[0])
financials = pd.concat([financials, previous_financials], axis = 0)
financials['PermID'] = financials['PermID'].astype(int)
print(financials.shape[0])
```
